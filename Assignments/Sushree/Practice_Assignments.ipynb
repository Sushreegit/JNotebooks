### 1. Create a new dataframe from df with just the policyID, point_latitude and point_logitude.

### 2. Sort the new dataframe in descending order of their eq_site_limit value.

```python
sort_df= new_df.sort_values('eq_site_limit', ascending= False)
sort_df.drop('eq_site_limit',axis=1)
```

### 3. Which state has the maximum total hu_site_deductible value?

```python
max_hu_site_deductible = df['hu_site_deductible'].max()
f1= df['hu_site_deductible'] == max_hu_site_deductible
print (df[f1].statecode)
```

### 4. What is the point_latitude range (min to max) for each type of construction?

```python
df.groupby('construction').point_latitude.agg(['min', 'max'])
```

### 5. Remove all duplicate records from the sub-dataframe containing the policyID and all \*_deductible variables.

```python
df.drop_duplicates(subset=['policyID','eq_site_deductible', 'hu_site_deductible','fl_site_deductible','fr_site_deductible'])
```

### 6. For every unique combination of line and construction find the mean *_deductible values.

```python
df.groupby(['line','construction']).agg({'eq_site_deductible':'mean','hu_site_deductible':'mean','fl_site_deductible':'mean','fr_site_deductible':'mean'})
```
